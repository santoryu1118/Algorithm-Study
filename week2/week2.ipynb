```python
#(3) - 2110 공유기 설치

import sys
N, C = map(int, sys.stdin.readline().split())
houses = [int(sys.stdin.readline()) for _ in range(N)]
houses.sort()

# 인덱스로 이진 탐색을 하는게 아님.
# 나무 자르기 처럼 거리값을 이진탐색으로 조절해주는거임

start = 1
end = houses[-1] - houses[0]
answer = 0

while start <= end:
    count = 1
    previous_house = houses[0]
    distance = (start + end) // 2
    for house in houses[1:]:
        if house - previous_house >= distance:
            count +=1
            previous_house = house

    # count 가 많으면 거리를 좀 늘려줘도 된다는 뜻이지
    if count >= C:  
        answer = max(answer, distance)
        start = distance + 1

    # count가 부족하면 거리를 좁혀줘야겟지
    elif count < C:
        end = distance -1 

print(answer) 
```
