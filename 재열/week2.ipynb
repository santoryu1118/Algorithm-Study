```python
#(16) - 2493 탑
import sys

N  = int(input())
tops = [int(x) for x in sys.stdin.read().split()]
tops_bool = [0]*N

# 첫번째 풀이 시간초과 뜸
for i, height in enumerate(tops):
    for j in range(i-1,-1,-1):
        if tops[j] >= height:
            tops_bool[i] = (j+1)
            break
for x in tops_bool:
    print(x, end=' ') 

# 두번쨰 풀이 
# index 저장해놓는 스택 
stack = [0]
sys.stdout.write('0 ')

for i in range(1,N):
    # 이렇게 pop을 해줄 시 가운데 불필요한 값들을 비교할 필요가 없어져서 시간 단축됨
    while len(stack)!=0 and tops[i] > tops[stack[-1]]:
        stack.pop()
    if len(stack) == 0:
        print ('0', end =' ')
    else:
        print(stack[-1]+1, end =' ')
    stack.append(i)
```

```python
#(17) -10000 원 영역
import sys

N = 4
points = []
for _ in range(N):
    x,r  = map(int, sys.stdin.readline().split())
    points.append(["{", x-r,0,0])
    points.append(["}", x+r, 0, 0])

points.sort(key =lambda x : (x[1]))
stack = []
answer = 1

for i in range(len(points)):
    if points[i][0] =="{":
        if stack:
            if stack[-1][1] == points[i][1]  or stack[-1][3] == points[i][1]:
                stack[-1][2] = 1
            else:
                stack[-1][2] = 0
        stack.append(points[i])
    else:
        half = stack.pop()
        if stack and stack[-1][2] == 1:
            stack[-1][3] = points[i][1]
        if half[2] == 1 and half[3] == points[i][1]:
            answer +=1
        answer +=1
print(answer)
```

```python
#(18) -2504 괄호의 값

import sys
string = sys.stdin.read().strip()
stack = []

def printsum(string):
    for i in string:
        tempnum = 0
        if i == '(' or i == '[':
            stack.append(i)

        elif i == ')':
            if '(' not in stack:
                return 0
            while stack[-1] != '(':
                k = stack.pop()
                if k == '[' or k == ']':
                    return 0
                else:
                    tempnum += k
            stack.pop()
            if tempnum == 0:
                stack.append(2)
            else:
                stack.append(2*tempnum)

        elif i == ']':
            if '[' not in stack:
                return 0
            while stack[-1] != '[':
                k = stack.pop()
                if k =='(' or k == ')':
                    return 0
                else:
                    tempnum += k
            stack.pop()
            if tempnum ==0:
                stack.append(3)
            else:
                stack.append(3*tempnum)

    if '(' in stack or ')' in stack or '[' in stack or ']' in stack:
        return 0

    return sum(stack)

print(printsum(string))
```

```python
#(20) - 18258 큐 2
import sys
from collections import deque
a = deque()
for i in range(int(sys.stdin.readline())):
    command = list(sys.stdin.readline().split())
    if command[0] == 'push':
        a.append(command[1])
    elif command[0] == 'pop':
        if len(a) == 0:
            print (-1)
        else: 
            print(a.popleft())
    elif command[0] == 'size':
        print(len(a))
    elif command[0] == 'empty':
        if len(a) == 0:
            print (1)
        else:
            print(0)
    elif command[0] == 'front':
        if len(a) == 0:
            print (-1)
        else:
            print(a[0])
    elif command[0] == 'back':
        if len(a) == 0:
            print (-1)
        else:
            print(a[-1])
```

```python
#(21) - 2164 카드 2
from collections import deque
N = deque(x+1 for x in range(int(input())))
while len(N) > 1:
    N.popleft()
    N.append(N.popleft())

for x in N:
    print(x)
```

```python
#(22) - 요세푸스 문제
import sys
n,k = map(int, sys.stdin.readline().split())
N = [x for x in range(1,n+1)]
answer = []
num = 0
for _ in range(n):
    num = (num + k-1)%len(N)
    answer.append(str(N.pop(num)))

print('<', ', '.join(answer), '>', sep ='')
```
