```python
#(18) -2504 괄호의 값

import sys
string = sys.stdin.read().strip()
stack = []

def printsum(string):
    for i in string:
        tempnum = 0
        if i == '(' or i == '[':
            stack.append(i)

        elif i == ')':
            if '(' not in stack:
                return 0
            while stack[-1] != '(':
                k = stack.pop()
                if k == '[' or k == ']':
                    return 0
                else:
                    tempnum += k
            stack.pop()
            if tempnum == 0:
                stack.append(2)
            else:
                stack.append(2*tempnum)

        elif i == ']':
            if '[' not in stack:
                return 0
            while stack[-1] != '[':
                k = stack.pop()
                if k =='(' or k == ')':
                    return 0
                else:
                    tempnum += k
            stack.pop()
            if tempnum ==0:
                stack.append(3)
            else:
                stack.append(3*tempnum)

    if '(' in stack or ')' in stack or '[' in stack or ']' in stack:
        return 0

    return sum(stack)

print(printsum(string))
```
